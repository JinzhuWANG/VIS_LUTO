### 1-4) Plot crosstab changes

```python
# HIDDEN
import pandas as pd

# read the cross table between start and end year
begin_end_df = files_selc.query('''year_types == "begin_end_year" and base_name.str.contains("crosstab-lumap")''').reset_index(drop=True)
begin_end_df = pd.read_csv(begin_end_df['path'][0], index_col=0)

# get the total area in the begin year
area_begin = begin_end_df['Total [km2]']

# remove the last row and column
begin_end_df = begin_end_df.iloc[:-1,:-1]

# divide the begain area so we get the percentage change
begin_end_df = begin_end_df.divide(area_begin, axis=0) * 100

# fill nan with 0
begin_end_df = begin_end_df.fillna(0)
```

```python
# convert the dataframe to long format
crosstab_lu_all = begin_end_df.reset_index(names='Land-use from')\
    .melt(id_vars='Land-use from', 
          value_vars=begin_end_df.columns, 
          var_name='Land-use to', 
          value_name='Area change (%)')
```

```python
# HIDE CODE
# plot the transition matrix using altair
import altair as alt

alt.Chart(crosstab_lu_all).mark_rect().encode(
    x='Land-use to:N',
    y='Land-use from:N',
    color=alt.Color('Area change (%):Q',
                    scale=alt.Scale(scheme='lightmulti'),
                    legend=alt.Legend(
                                    title="Area change (%)",
                                    orient='none',
                                    legendX=420, legendY=100,
                                    direction='vertical',
                                    titleAnchor='start')),
    tooltip=['Land-use from','Land-use to','Area change (%)']
).properties(
    width=400,
    height=400
)
```
