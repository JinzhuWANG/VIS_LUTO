```python
import re
from tools import get_all_files

from tools.Plot_AREA import get_AREA_lu, plot_area_lu,\
                            get_AREA_lm, plot_area_lm,\
                            get_AREA_am, plot_area_am_total, plot_area_am_lu
```

```python
# Get the root directory of the data
data_root = '../2023_11_15__10_08_15_hard_mincost_RF30_P1e4_Won_Gon'

# Get all output files
files = get_all_files(data_root)
```

```python
# Select files based on criteria
files_selc = files.query('year <= 2050')
print(files_selc['catetory'].unique(),'\n')

files_selc
```

## 1 ) Get the total area (km2) change

```python
Area_df = files_selc.query('catetory == "cross_table"').reset_index(drop=True)
Area_df
```

```python
Area_df['base_name'].unique()
```

### 1-1) Plot land use areas between years

```python
crosstab_lu = Area_df.query('base_name == "crosstab-lumap"').reset_index(drop=True)
crosstab_lu.head(3)
```

```python
lu_area = get_AREA_lu(crosstab_lu)
plot_area_lu(lu_area)
```

### 1-2) Plot the total area (km2) for lm

```python
crosstab_lm = Area_df.query('base_name == "crosstab-lmmap"').reset_index(drop=True)
crosstab_lm.head(3)
```

```python
lm_area = get_AREA_lm(crosstab_lm)
plot_area_lm(lm_area)
```

### 1-3) Plot the area change of agricultural managements

```python
switches_am = Area_df.query('base_name.str.contains(r"switches.*amstat.*", regex=True)').reset_index(drop=True)
switches_am.head(3)
```

```python
am_area_km2 = get_AREA_am(switches_am)
am_area_km2[['Land use','Agricultural management']] = am_area_km2['Land use']\
                                                     .apply(lambda x: re.findall(r'(.*) \((.*)\)',x)[0]).tolist()

plot_area_am_lu(am_area_km2) 
```

```python
am_area_km2_total = am_area_km2.groupby(['Year','Agricultural management']).sum(numeric_only=True).reset_index()
plot_area_am_total(am_area_km2_total)
```
