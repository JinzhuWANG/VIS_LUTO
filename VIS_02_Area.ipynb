```python
# HIDDEN
from PARAMETERS import AG_LANDUSE, NON_AG_LANDUSE
import pandas as pd
import numpy as np

total_area = lu_area
all_lucc = AG_LANDUSE + NON_AG_LANDUSE
crosstab_lu = crosstab_lu.sort_values(by=['year']).reset_index(drop=True)

year_start = 2010
year_end = 2050
```

```python
# HIDDEN
dfs = []
for _,row in crosstab_lu.iterrows():
    year = row['year']

    # get the area dataframe and remove the last row and column
    df = pd.read_csv(row['path'],index_col=0).iloc[:-1,:-1]
    # reindex teh dataframe to avoid missing land use classes
    df = df.reindex(all_lucc, axis=0).reindex(all_lucc, axis=1)
    # move the index to a column
    df.index = pd.MultiIndex.from_product([[year],df.index])
    # add the year column
    df.insert(0,'year',year)

    dfs.append(df)

# get the transition matrix for all years
crosstab_lu_all = pd.concat(dfs)
crosstab_lu_all = crosstab_lu_all\
            .groupby(level=1,axis=0)\
            .sum(numeric_only=True)\
            .drop(columns=['year'])

crosstab_lu_all = crosstab_lu_all.reindex(columns=crosstab_lu_all.index)

# fill the diagonal with the last year's values
crosstab_lu_all_val = crosstab_lu_all.values
np.fill_diagonal(crosstab_lu_all_val, np.diag(df.drop(columns=['year'])))
# np.fill_diagonal(crosstab_lu_all_val, 0)

crosstab_lu_all[:] = crosstab_lu_all_val

crosstab_lu_all = crosstab_lu_all.reset_index(names='Land-use from')

# convert the dataframe to long format
crosstab_lu_all = crosstab_lu_all\
    .melt(id_vars='Land-use from', 
          value_vars=crosstab_lu_all.columns[1:], 
          var_name='Land-use to', 
          value_name='Area (km2)')\
    .reset_index(names='from')
```

```python
# HIDE CODE
# plot the transition matrix using altair
import altair as alt

alt.Chart(crosstab_lu_all.reset_index()).mark_rect().encode(
    x='Land-use from:N',
    y='Land-use to:N',
    color='Area (km2):Q',
    tooltip=['Land-use from','Land-use to','Area (km2)']
).properties(
    width=400,
    height=400
).interactive()
```
